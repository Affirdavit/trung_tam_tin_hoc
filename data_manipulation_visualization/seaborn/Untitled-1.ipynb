```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Box Plot

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
# Ve box plot cho tap du kieu tips voi truc x la sex va truc y la total_bill
sns.boxplot(data=tips, x='sex', y='total_bill')
plt.show()
```

```python
q1 = np.quantile(tips[tips['sex'] == 'Male']['total_bill'], 0.25)
q3 = np.quantile(tips[tips['sex'] == 'Male']['total_bill'], 0.75)
```

```python
iqr = q3 - q1
upper_men = q3 + 1.5 * iqr
lower_men = q1 - 1.5 * iqr
```

```python
djia = pd.read_csv('djia.csv')
djia.columns = djia.columns.str.strip()
djia['Date'] = pd.to_datetime(djia['Date'], format='%m/%d/%y')
djia['Month'] = djia['Date'].dt.month_name()
djia['Month_num'] = djia['Date'].dt.month
djia = djia.sort_values(by='Month_num')
djia.head()
```

```python
plt.figure(figsize=(10,4))
sns.boxplot(data=djia, x='Month', y='Close')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 4))
sns.violinplot(data=djia, x='Month', y='Close')
plt.xticks(rotation=45)
plt.show()
```

```python
sns.regplot(data=djia, x='Open', y='Close')
plt.show()
```

```python
sns.scatterplot(data=djia, x='Open', y='Close')
plt.show()
```

```python
sns.residplot(data=djia, x='Open', y='Close')
plt.show()
```

```python
plt.figure(figsize=(10,4))
sns.barplot(data=djia, x='Month', y='Close')
plt.show()
```

```python
sns.barplot(data=tips, x='day', y='total_bill', errorbar=None, hue='sex')
```

```python
plt.figure(figsize=(10,6))
sns.countplot(data=djia, x='Month')
plt.grid(axis='y')
plt.show()
```
