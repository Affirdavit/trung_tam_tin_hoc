```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme()
```

```python
emp_df = pd.read_csv('employees_full.csv')
```

```python
emp_df.head()
```

```python
cnt_ser = emp_df['DEPARTMENT_NAME'].value_counts()
cnt_ser.plot.barh()
```

```python
salary_ser = emp_df.groupby('DEPARTMENT_NAME')['SALARY'].mean().sort_values(ascending=False)
```

```python
salary_ser.plot.bar(figsize = (8,5), grid=True)
plt.xlabel('Department Name')
plt.ylabel('Average Salary')
plt.title('Average of salary per department')
```

```python
ser = emp_df.groupby('DEPARTMENT_NAME')['SALARY'].agg(['min', 'max'])
```

```python
ser.plot.bar(figsize=(10, 6))
```
