```python
djia = pd.read_csv('djia.csv')
```

```python
djia
```

```python
djia['Date'] == pd.to_datetime(djia['Date'])
```

```python
djia.info()
```

```python
djia.columns = djia.columns.str.strip()
```

```python
djia.info()
```

```python
djia.set_index('Date', inplace=True)
djia
```

```python
djia.reset_index(inplace=True)
```

```python
djia['Date'] = pd.to_datetime(djia['Date'], format='%m/%d/%y')
```

```python
djia
```

```python
djia.plot(y=['Close'], figsize=(16, 6))
```

```python
plt.hist(djia['Close'], bins=20, color='blue')
plt.xlabel('Close price')
plt.ylabel('Frequency')
plt.title('Frequency of close price')
plt.show()
```

```python
plt.figure(figsize=(10, 4))
plt.plot(djia['Date'], djia['Close'], label='Close')
plt.plot(djia['Date'], djia['Open'], label='Open')
plt.legend()
plt.xticks(rotation=60)
plt.show()
```

```python
djia['month'] = djia['Date'].dt.month_name()
djia
```

```python
djia['year'] = djia['Date'].dt.year
```

```python
djia['month_num'] = djia['Date'].dt.month
```

```python
djia_monthly_mean = djia.groupby(['month', 'month_num'], as_index=False)['Close'].mean()
djia_monthly_mean.sort_values(by='month_num', ascending=True, inplace=True)
djia_monthly_mean
```

```python
plt.figure(figsize=(10,4))
plt.bar(x=djia_monthly_mean['month'], height = djia_monthly_mean['Close'])
plt.xticks(rotation=60)
plt.show()
```
